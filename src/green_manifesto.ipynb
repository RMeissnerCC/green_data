```python
import requests as requests

url = "https://cms.gruene.de/uploads/documents/20210311_Grundsatzprogramm_EN.pdf"
filename = "green_manifesto.pdf"

import textract

response = requests.get(url)

# Write content in pdf file
with open(filename, 'wb') as pdf:
    pdf.write(response.content)
    pdf.close()
```

```python
import pdfminer
PDF_read = textract.process(filename, method='pdfminer')
```

```python
text = PDF_read.decode()
```

```python
print(set(text.replace("\n", "").split(" ")))
```

```python
import string

# Create an empty dictionary
d = dict()

# Loop through each line of the file
for line in text.split("\n"):
    # Remove the leading spaces and newline character
    line = line.strip()

    # Convert the characters in line to
    # lowercase to avoid case mismatch
    line = line.lower()

    # Remove the punctuation marks from the line
    line = line.translate(line.maketrans(",", " ", string.punctuation))

    # Split the line into words
    words = line.split(" ")

    # Iterate over each word in line
    for word in words:
        # Check if the word is already in dictionary
        if word in d:
            # Increment count of word by 1
            d[word] = d[word] + 1
        else:
            # Add the word to dictionary with count 1
            d[word] = 1

# Print the contents of dictionary
for key in list(d.keys()):
    print(key, ":", d[key])
```
